## **Building Amazing word cloud**

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
```

```python
from transformers import pipeline
model = pipeline("text-generation", model = "gpt2")
```

```python
sentence = model("Hi, My name is John Cena, I am here", do_sample=True, top_k=50, temperature=0.9, max_length=100)

# for i in sentence:
#   print(i["generated_text"])
```

```python
sentences = (i["generated_text"] for i in sentence)
s=list(sentences)
s
```

```python
wordcloud = WordCloud(width = 800, height = 800,background_color ='black',min_font_size = 7).generate(s[0])
 
# plot the WordCloud image                       
plt.figure(figsize = (7, 7), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.savefig("wordcloud")
plt.show()
```
